<font color=green><h3><b>$\alpha$. Split observations into train and test sets</b><h3>

```python
%%time

tX_in = tX.apply(pd.to_numeric, errors='coerce')
vX_in = vX.apply(pd.to_numeric, errors='coerce')

tX_in = tX_in.fillna(tX_in.median())
vX_in = vX_in.fillna(tX_in.median())

tY_in = tY.astype(int)
```

<font color=green><h3><b>$\beta$. Build and fit a model</b><h3>

```python
from sklearn.ensemble import ExtraTreesClassifier

m = ExtraTreesClassifier(
    n_estimators=120,
    max_depth=None,
    min_samples_leaf=1,
    max_features='sqrt',
    random_state=0,
    n_jobs=-1
)

m.fit(tX_in, tY_in)
```

<font color=green><h3><b>$\gamma$. Make predictions</b><h3>

```python
pY = pd.DataFrame(m.predict(vX_in).astype(int), index=vX.index + 1, columns=['y'])
ToCSV(pY, 'submission')
print('Observed  y distribution: ', list((tXY.y.value_counts()/len(tXY)).round(3)))                # distribution of training target level
print('Predicted y distribution: ', list((pY.round(0).astype(int).value_counts()/len(pY)).round(3))) # distribution of test target level
```

## Task 1. Preprocessing Ideas

1. **Title**: Median Imputation for Stable Tree-Based Classification

2. **Idea Discovery**:
   During exploratory analysis of the dataset, we found that some feature values were missing. Since tree-based ensemble methods require complete numerical inputs, we investigated simple imputation strategies. We selected median imputation because it is robust to outliers and can be applied efficiently to a large dataset.

3. **Finding's Importance**:
   Missing values may prevent model training or introduce instability in the learned decision rules. Median imputation preserves the overall feature distributions better than replacing missing values with arbitrary constants and is less sensitive to extreme observations than mean imputation.

4. **Experiment Setup**:
   All features were converted to numeric format using `pd.to_numeric(errors='coerce')`. Missing values in the training set were replaced with feature medians, and the same medians were used to fill missing values in the test set. The preprocessing pipeline was evaluated together with several tree-based ensemble models. Runtime was also considered because the notebook had to satisfy a strict execution limit.

5. **Results**:
   Median imputation produced a stable preprocessing pipeline with low computational overhead. The resulting feature matrix was successfully used by ensemble classifiers without increasing runtime significantly. This preprocessing approach was retained for all subsequent experiments.

## Task 2. Modeling Ideas

1. **Title**: Replacing Random Forest with Extra Trees Under Runtime Constraints

2. **Idea Discovery**:
   The baseline solution used a single Decision Tree classifier, which produced limited predictive performance. We experimented with stronger ensemble methods taught in the course, including Random Forest and Extra Trees. Initial experiments showed that Random Forest improved predictive power but frequently exceeded the competition runtime limit when the number of trees was increased.

3. **Finding's Importance**:
   The competition imposed a runtime limit of 60 seconds. Therefore, model selection required balancing predictive performance and computational efficiency. Extra Trees are known to reduce variance through stronger randomization while often training faster than Random Forest because split thresholds are selected randomly rather than through exhaustive search.

4. **Experiment Setup**:
   We compared several ensemble configurations using the same preprocessing pipeline. Random Forest models with different numbers of trees and depths were tested, but many configurations exceeded the runtime budget. We then evaluated Extra Trees classifiers with different hyperparameters. The final model used:

* `n_estimators = 120`
* `max_depth = None`
* `min_samples_leaf = 1`
* `max_features = sqrt`
* `random_state = 0`
* `n_jobs = -1`

The evaluation criterion was the competition leaderboard score together with compliance with the runtime limit.

5. **Results**:
   The final Extra Trees model completed training in approximately 46 seconds, satisfying the runtime constraint. It achieved a leaderboard score of approximately **0.982**, outperforming the baseline Decision Tree and providing a better trade-off between accuracy and execution time than the tested Random Forest configurations. Therefore, Extra Trees was selected as the final model.
